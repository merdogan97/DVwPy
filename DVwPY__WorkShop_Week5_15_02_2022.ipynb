```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

# **Use below dataframe for plots**

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

**Bar plots**

# **1. Set darkgrid style from seaborn**

```python
sns.set(style="darkgrid")
```

# **2. Make a bar plot**

```python
sns.barplot(x="income", y="expense", data=df);
```

```python
fig, ax = plt.subplots(figsize=(20,10))
p = np.arange(len(index))
width = 0.20
ax.bar(p - width/2, income, width, label="income")
ax.bar(p + width/2, expense, width, label="expense")    
ax.set_xticks(p)
ax.set_xticklabels(index)
plt.legend()
for p in ax.patches:
  ax.annotate(p.get_height(),(p.get_x()+0.06,p.get_height()+1 ))
plt.show()
```

## **3. Stack the bars**

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.bar(index, income, label="income")
ax.bar(index, expense, label="expense")
plt.legend()
for p in ax.patches:
  ax.annotate(p.get_height(),(p.get_x()+0.3,p.get_height()+1 ))
plt.show()
```

# **4. Rotate the labels and set figsize**

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.bar(index, expense, label="expense")
ax.bar(index, income, label="income")
plt.legend()
for p in ax.patches:
  ax.annotate(p.get_height(),(p.get_x()+0.3,p.get_height()+1 ))
plt.show()
```
