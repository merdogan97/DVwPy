# **Functional Method**

```python

x = np.linspace(0,10,10) 
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)  # karekök

plt.figure()
plt.subplot(2,2,1);  # 1. grafik
plt.plot(x,y1, 'r+');
plt.title('y1=x');

plt.subplot(2,2,2)    # 2. grafik
plt.plot(x,y2, 'g--');
plt.title('$y_2=x^2$');   #   $ italik göstermek için

plt.subplot(2,2,3);   # 3. grafik
plt.plot(x,y3,'b^');
plt.title('$y_3=x^3$');

plt.subplot(2,2,4);
plt.plot(x,y1, 'ks');
plt.title('$y_4=\sqrt{x}$');
```

```python
x = np.linspace(0,10,10) 
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)  # karekök

# plt.figure(figsize=(14,10))  büyütmek için figsize
plt.subplot(4,1,1)   # 4 satır ve 1 sütun
plt.plot(x,y1, 'ro');
plt.title('$y_1=x$');

plt.subplot(4,1,2)
plt.plot(x,y2, 'g--');
plt.title('$y_2=x^2$');

plt.subplot(4,1,3)
plt.plot(x,y3,'b^');
plt.title('$y_3=x^3$');

plt.subplot(4,1,4)
plt.plot(x,y4, 'ks');
plt.title('$y_4=\sqrt{x}$');
```

# **Pie Chart**

```python

# matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None, 
# pctdistance=0.6, labeldistance=1.1, startangle=0, radius=1, data=None)
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1, autopct='%0.f%%') # 0. ise decimal yok ===>> 0.1 ise 1 decimal
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.1f%%')
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.1f%%', shadow=True, explode=[0,0.4,0,0,0.8], startangle=180)
plt.show()
```
